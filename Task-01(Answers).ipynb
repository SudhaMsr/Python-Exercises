### Question 6

A year is a leap year if:

    Rule 1:
    Divisible by 400 → Leap year
    
    Rule 2:
    Divisible by 4 AND NOT divisible by 100 → Leap year
    
    Otherwise:
    Not a leap year

```python
#Question 6--Write a function
n= int(input())

if n %400 ==0:
    print(True)
elif n%4 == 0 and n%100 !=0:
    print(True)
else:
    print(False)
```

```python
def is_leap(year):
    return (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0)

year = int(input())
print(is_leap(year))
```
